```python
import pandas as pd
```

Loading the population dataset and removing extra columns, leaving only columns: 
- Time - renamed to Year
- Age
- Sex 
- Value - renamed to Population

```python
df = pd.read_csv('..\dataset\population_dataset.csv')

df = df.drop(columns=['IndicatorId', 
                      'IndicatorName', 
                      'IndicatorShortName', 
                      'Source', 
                      'SourceYear', 
                      'Author', 
                      'LocationId', 
                      'Location', 
                      'Iso2', 
                      'Iso3', 
                      'TimeId', 
                      'VariantId', 
                      'Variant', 
                      'SexId', 
                      'AgeId',
                      'AgeStart', 
                      'AgeEnd', 
                      'Category', 
                      'CategoryId', 
                      'EstimateTypeId', 
                      'EstimateType', 
                      'EstimateMethodId', 
                      'EstimateMethod'])

df = df.rename(columns={'Value': 'Population',
                        'Time': 'Year'})

df.head()
```

Current age groups are: 0, 0-14, 15-64, 65+, 90+.
Desired age groups defined by the task are: 0, 1-14, 15-64, 65-89 and 90+. In the following code, defined age groups are derived from what the UN Population Dataset provided us.

```python
cleaned_rows = []

for year in df['Year'].unique():
    for sex in df['Sex'].unique():

        subset = df[(df['Year'] == year) & (df['Sex'] == sex)]

        pop_0 = subset.loc[subset['Age'] == '0', 'Population'].values[0]
        pop_0_14 = subset.loc[subset['Age'] == '0-14', 'Population'].values[0]
        pop_15_64 = subset.loc[subset['Age'] == '15-64', 'Population'].values[0]
        pop_65_plus = subset.loc[subset['Age'] == '65+', 'Population'].values[0]
        pop_90_plus = subset.loc[subset['Age'] == '90+', 'Population'].values[0]

        pop_1_14 = pop_0_14 - pop_0
        pop_65_89 = pop_65_plus - pop_90_plus

        cleaned_rows.append([year, "0", sex, pop_0])
        cleaned_rows.append([year, "1-14", sex, pop_1_14])
        cleaned_rows.append([year, "15-64", sex, pop_15_64])
        cleaned_rows.append([year, "65-89", sex, pop_65_89])
        cleaned_rows.append([year, "90+", sex, pop_90_plus])

df = pd.DataFrame(cleaned_rows, columns=['Year', 'Age', 'Sex', 'Population'])

df.head(20)
```

Saving the cleaned dataset to a new CSV file.
