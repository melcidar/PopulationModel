```python
import pandas as pd
```

In this notebook, general parameters table is created. This table will contain only Total Fertility Rate, Crude Rate of Net Migration, together with probability of a baby being female or male (which are constants).
Furthermore, Crude Rate of Net Migration is given as the number per 1000. Therefore, this number will be divided by 1000 and saved in column Net Migration Rate.

```python
df = pd.read_csv('../dataset/general_parameters_dataset.csv')

df = df.drop(columns=['IndicatorId', 
                      'IndicatorShortName', 
                      'Source', 
                      'SourceYear', 
                      'Author', 
                      'LocationId', 
                      'Location', 
                      'Iso2', 
                      'Iso3', 
                      'TimeId', 
                      'VariantId', 
                      'Variant', 
                      'SexId', 
                      'Sex',
                      'AgeId',
                      'AgeStart', 
                      'AgeEnd', 
                      'Age',
                      'Category', 
                      'CategoryId', 
                      'EstimateTypeId', 
                      'EstimateType', 
                      'EstimateMethodId', 
                      'EstimateMethod'])

df = df.rename(columns={'Time': 'Year',
                        'IndicatorName': 'Indicator'})

df.head()
```

```python
df = df.pivot_table(
    index="Year",
    columns="Indicator",
    values="Value",
    aggfunc="first"
).reset_index()

df['p_male'] = 0.5134
df['p_female'] = 0.4866

df['Crude rate of net migration'] = df['Crude rate of net migration'] / 1000
df = df.rename(columns={'Crude rate of net migration': 'Net Migration Rate'})
print(df)
```

```python
df.to_csv('../dataset/cleaned_general_parameters_dataset.csv', index=False)
```
