```python
import pandas as pd

raw_un_rates = pd.read_csv("../dataset/UN_future_parameters.csv")

print(raw_un_rates.columns)
print(raw_un_rates.head())
print(raw_un_rates.columns.tolist())
print(raw_un_rates["IndicatorName"].unique())
```

```python
un_rates = raw_un_rates[["Time", "IndicatorName", "Value"]].copy()

un_rates = un_rates.rename(columns={
    "Time": "Year",
    "IndicatorName": "Indicator",
    "Value": "Value"
})

un_rates["Value"] = pd.to_numeric(un_rates["Value"], errors="coerce")

un_rates_pivot = un_rates.pivot_table(
    index="Year",
    columns="Indicator",
    values="Value",
    aggfunc="first"
).reset_index()

un_rates_pivot.columns.name = None

print(un_rates_pivot.head())
print(un_rates_pivot.columns.tolist())
```

```python
un_rates_pivot = un_rates_pivot.rename(columns={
    "Crude birth rate": "UN birth rate per 1000",
    "Crude death rate": "UN death rate per 1000",
    "Total net-migration": "UN total net migration"
})

un_rates_pivot.head()
```

```python
un_rates_pivot.to_csv(
    "../dataset/cleaned_un_demographic_rates_2020_2030.csv",
    index=False
)
```
