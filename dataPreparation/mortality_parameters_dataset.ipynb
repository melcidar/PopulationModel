```python
import pandas as pd
```

In the following code, parameters related to mortality is prepared.
The mortality data was available in detailed age groups such as 1-4, 5-9, 10-14, and so on. These groups were mapped to the broader age groups used in the model: 0, 1-14, 15-64, 65-89, and 90+. For each year and sex, the mortality rate of each broader group was calculated as the average of the corresponding detailed age-specific mortality rates.

```python
df = pd.read_csv("../dataset/mortality_parameters_dataset.csv")


df = df.rename(columns={
    "Time": "Year",
    "Sex": "Sex",
    "Age": "age_original",
    "Value": "Mortality Rate"
})

df = df[["Year", "Sex", "age_original", "Mortality Rate"]]

age_mapping = {
    "0": "0",

    "1-4": "1-14",
    "5-9": "1-14",
    "10-14": "1-14",

    "15-19": "15-64",
    "20-24": "15-64",
    "25-29": "15-64",
    "30-34": "15-64",
    "35-39": "15-64",
    "40-44": "15-64",
    "45-49": "15-64",
    "50-54": "15-64",
    "55-59": "15-64",
    "60-64": "15-64",

    "65-69": "65-89",
    "70-74": "65-89",
    "75-79": "65-89",
    "80-84": "65-89",
    "85-89": "65-89",

    "90-94": "90+",
    "95-99": "90+",
    "100+": "90+"
}

df["Age"] = df["age_original"].replace(age_mapping)

df = df[df["Age"].isin(["0", "1-14", "15-64", "65-89", "90+"])]


df = (
    df
    .groupby(["Year", "Age", "Sex"], as_index=False)
    .agg(**{"Mortality Rate": ("Mortality Rate", "mean")})
)

df.to_csv("../dataset/mortality_parameters_dataset_cleaned.csv", index=False)

df.head(20)
```
